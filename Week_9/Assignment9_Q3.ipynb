### UNHCR Immigration Statistics API

```python
import requests
import pandas as pd
```

### UNHCR Asylum seekers for France from January 2015

```python
url = 'http://popdata.unhcr.org/api/stats/asylum_seekers_monthly.json?year=2015&month=1&coa=FRA'
data = requests.get(url)
as_json = data.json()
df = pd.DataFrame(as_json)
```

```python
df.sort_values(by='value', ascending = False)
```

### France recieved the most immigrants in January 2015 from Serbia and Kosovo. 

### What were the 5 countries that contributed the most immigrants in January of 2015? Serbia and Kosovo, Democratic Republic of Congo, Russia, China, Haiti

```python
df_1 = df.loc[df['value'] == 1]
pd.set_option('display.max_rows', 100)
df_1
```

### From what countries did only 1 immigrant arrive from France in January of 2015? see list above - there are 83
