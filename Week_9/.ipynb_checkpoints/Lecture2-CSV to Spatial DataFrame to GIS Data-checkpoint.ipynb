### Filter data - 0 points and by data

```python
crime_csv['XCoord']!= 0
```

```python
filtered_crime_csv = crime_csv[crime_csv['XCoord'] != 0]
filtered_crime_csv
```

### lost about 100 points

```python
filtered_crime_csv = crime_csv[(crime_csv['XCoord'] != 0) | (crime_csv['YCoord'] != 0)]
len(filtered_crime_csv)
```

### ^ just checking (X and Y coordinate both 0)

### Filter by date

```python
new_crim_df = filtered_crime_csv[filtered_crime_csv.DateOccur >= '07/01/2018']
len(new_crim_df)
```

## Turn this info into feature class using Spatial DataFrame

```python
import arcgis
from arcgis.features import SpatialDataFrame
```

```python
### wkid = well-known ID (state plane maybe)
wkid = 102696
crime_coords = new_crim_df.iloc[:,18:20].values.tolist()
crime_coords
```

### turn pandas dataframe to spatial dataframe

```python
sdf = SpatialDataFrame(new_crim_df,geometry=[arcgis.geometry.Geometry({'x':r[0], 'y':r[1], 
                    'spatialReference':{'wkid':wkid}}) for r in crime_coords])
```

```python
sdf.head()
```

### spatial data frame to feature class

```python
from arcgis.gis import GIS
from arcgis.geoenrichment import *

gis = GIS("http://slustl.maps.arcgis.com/home", "laurenlovato_slustl")
```

```python
stl_map = gis.map('St. Louis, MO')
stl_map
```

### need to remove NA values from SDF geometry (SHAPE) and reset dataframe indices (so they start at 0)

```python
sdf = sdf[sdf.SHAPE.notna()].copy()
```

```python
sdf.reset_index(inplace=True, drop=True)
```

```python
crime_map = gis.content.import_data(sdf, title='STL July Crime LL2')
```

```python
stl_map.add_layer(crime_map, {"renderer":"HeatmapRenderer",
               "opacity": 0.75})
```
