###  Earthquakes from the past week from USGS Earthquakes Hazard Program

```python
import arcgis
from arcgis.features import SpatialDataFrame 
import pandas as pd
```

```python
eqw_csv = pd.read_csv("all_week.CSV")
eqw_csv.head()
```

```python
eqw2_csv = pd.read_csv("all_week_aug_13_20.CSV")
eqw2_csv.head()
```

```python
df1 = pd.DataFrame(eqw_csv)
df2 = pd.DataFrame(eqw2_csv)
```

```python
comb_df = df1.append(ignore_index=True, other=df2)
```

```python
locations = comb_df.iloc[:,1:3].values.tolist()
```

```python
sdf = SpatialDataFrame(comb_df,geometry=[arcgis.geometry.Geometry({'x':r[1], 'y':r[0], 
                    'spatialReference':{'wkid':4326}}) for r in locations])
```
