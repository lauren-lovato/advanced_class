#### Create a Webmap that shows the magnitude 5 earthquake locations from USGS Earthquakes Hazard Program from Past Day 

```python
import arcgis
from arcgis.features import SpatialDataFrame 
```

```python
import pandas as pd
```

```python
earthquake_csv = pd.read_csv("all_day.CSV")
earthquake_csv.head()
```

```python
df = pd.DataFrame(earthquake_csv)
```

```python
new_df = df.drop(columns= ["nst", "gap", "dmin", "rms", "net"])
```

```python
new_df.head()
```

```python
locations = new_df.iloc[:,1:3].values.tolist()
```

```python
locations
```

```python
sdf = SpatialDataFrame(new_df,geometry=[arcgis.geometry.Geometry({'x':r[1], 'y':r[0], 
                    'spatialReference':{'wkid':4326}}) for r in locations])
```

```python
sdf.head()
```

```python
sdf1 = sdf[sdf['mag'] < 2.5]
len(sdf1)
```

```python
sdf2 = sdf[(sdf['mag']>2.5) & (sdf['mag'] < 4)]
len(sdf2)
```

```python
sdf3 = sdf[(sdf['mag']>4) & (sdf['mag'] < 5)]
len(sdf3)
```

```python
sdf4 = sdf[sdf['mag'] > 5]
len(sdf4)
```

```python
sdf4.reset_index(inplace=True, drop=True)
```

```python
from arcgis.gis import GIS
gis = GIS("http://slustl.maps.arcgis.com/home", "laurenlovato_slustl")
```

```python
eq = gis.content.import_data(sdf4, title='Magnitude Greater than 5 LL')
eq
```

#### gave me an error but created Feature Service on ArcGIS Online

```python
map1 = gis.map()
map1
```

```python
map1.add_layer(eq, {"renderer":"HeatmapRenderer",
               "opacity": 0.75})
```

```python
webmap_properties = {'title':'Magnitude Greater than 5 Webmap LL',
                    'snippet': 'Earthquakes with magnitude greater than 5 from Nov. 6',
                    'tags':['earthquake', 'USGS']}

webmap_item = map1.save(webmap_properties)
webmap_item
```
