```python
from arcgis.gis import GIS
import json
import pandas as pd
gis = GIS("http://slustl.maps.arcgis.com/home", "laurenlovato_slustl")
```

```python
df=pd.read_json('http://popdata.unhcr.org/api/stats/resettlement.json?year=2017&country_of_asylum=GBR')
df
```

```python
fc = gis.content.import_data(df, {"Address" : "country_of_origin_en"})
```

```python
item_properties = {"title": "GBR Asylum 2017", "tags": "GBR, refugees",
                  "text": json.dumps({"featureCollection": {"layers": [dict(fc.layer)]}}),
    "type": "Feature Collection"}
item = gis.content.add(item_properties)
```
