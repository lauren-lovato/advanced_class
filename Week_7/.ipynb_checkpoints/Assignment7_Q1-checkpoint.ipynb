### Using Python, publish the rainfall data contained in the national_rainfall_data.zip file. Be sure to verify that the service publishes by calling display(published_shp) and then calling add_layer(...) to add the layer to a map.

```python
from arcgis.gis import GIS
gis = GIS("http://slustl.maps.arcgis.com/home", "laurenlovato_slustl")
```

```python
import os

data_oldname = "national_rainfall_data_gjb.zip"
data = "national_rainfaill_data_LL.zip"

os.rename(data_oldname, data)
```

```python
rainfall = gis.content.add({"title": "National Rainfall Data LL", "tags":["rainfall", "United States"]}, data)
```

```python
published_rainfall = rainfall.publish()
```

```python
published_rainfall
```

```python
rainfall_map = gis.map("United States")
rainfall_map
```

```python
rainfall_map.add_layer(published_rainfall)
```
