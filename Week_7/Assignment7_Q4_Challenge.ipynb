```python
from arcgis.gis import GIS
gis = GIS("http://slustl.maps.arcgis.com/home", "laurenlovato_slustl")
```

```python
crime_search = gis.content.search("St. Louis Crime LL", "feature layer", max_items=1)
crime_item = crime_search[0]
crime_item
```

```python
ward_search = gis.content.search("St. Louis Neighborhoods LL", "feature layer", max_items=1)
ward_item = ward_search[0]
ward_item
```

```python
from arcgis.features import summarize_data
crime_summary = summarize_data.aggregate_points(point_layer = crime_item, 
                                                output_name="stl_crime_by_ward_ll",
                                                polygon_layer = ward_item,
                                                keep_boundaries_with_no_points=False)
```

```python
crime_summary
```

```python
crime_summary.layers[0]
```

```python
crime_fc = crime_summary.layers[0]
crime_fset = crime_fc.query()
```

```python
crime_fset
```

```python
crime_df = crime_fset.df
crime_df.head()
```

```python
%matplotlib inline
crime_df.plot('NHD_NAME','Point_Count', kind='bar', figsize=(20,10))
```
