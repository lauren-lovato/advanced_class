### Simple renderer

```python
simple_renderer = {
  "renderer": {
    "type": "simple",
    "symbol": {
      "color": [
        0,
        0,
        128,
        128
      ],
      "size": 15,
      "angle": 0,
      "xoffset": 0,
      "yoffset": 0,
      "type": "esriSMS",
      "style": "esriSMSCircle",
      "outline": {
        "color": [
          0,
          0,
          128,
          255
        ],
        "width": 0.99975,
        "type": "esriSLS",
        "style": "esriSLSSolid"
      }
    }
  }
}
```

### Heatmap renderer

```python
heatmap_renderer = {"renderer":"HeatmapRenderer", "opacity": 0.75}
```

### Classed size renderer

```python
classed_size_renderer = { "renderer": "ClassedSizeRenderer", "field_name":"TOTAL_FATALITIES"}
```

### Classed color renderer

```python
classed_color_renderer = {"renderer":"ClassedColorRenderer","field_name": "SHOT_MADE_FLAG"}
```

```python
from arcgis.gis import GIS
```

```python
gis = GIS ("http://slustl.maps.arcgis.com/home", "laurenlovato_slustl")
```

```python
acled = gis.content.search("ACLED", item_type="feature service", outside_org=True)
```

```python
acled
```

```python
acled[0]
```

```python
map1 = gis.map("Africa")
map1
```

```python
map1.add_layer(acled[0], simple_renderer)
```

```python
map2 = gis.map("Africa")
map2
```

```python
map2.add_layer(acled[0], heatmap_renderer)
```

```python
map3 = gis.map("Lagos, Nigeria")
map3
```

```python
map3.add_layer(acled[0], classed_size_renderer)
```

```python
classed_color_renderer = {"renderer":"ClassedColorRenderer","field_name": "EVENT_TYPE"}
```

```python
map4= gis.map("Lagos, Nigeria")
map4
```

```python
map4.add_layer(acled[0], classed_color_renderer)
```

```python
webmap_item_properties = {"title":"LL ACLED Simple Renderer",
              "snippet": "ACLED Data",
              "tags":['Acled', 'Africa']}
map1.save(webmap_item_properties)
```
